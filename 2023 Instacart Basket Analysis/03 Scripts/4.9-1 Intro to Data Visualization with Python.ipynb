## 4.9-1.4 Wrangle the data so that it follows consistent logic.

## 4.9-1.5 Complete the fundamental data quality and consistency checks.

```python
# Check stats
customers.describe()
```

```python
# Check data types
customers.dtypes
```

```python
# Check for mixed_type data
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col, 'mixed')
  else: print(col, 'NOT mixed')
```

```python
# Check for missing values
customers.isnull().sum()
```

```python
# Check shape
customers.shape
```

```python
# Check for duplicates
customers_dups = customers[customers.duplicated()]
```

```python
# Check output
customers_dups
```
