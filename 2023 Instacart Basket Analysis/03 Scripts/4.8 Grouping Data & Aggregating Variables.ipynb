## 4.8.4 Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Create max_order column using transform()
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check output
ords_prods_merged.head(15)
```

```python
# Creating loyalty_flag column using loc()
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print frequency of loyalty_flag column
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check output of user_id, loyalty_flag and order_number columns
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(20)
```

## 4.8.5 The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Performing multiple aggregations
ords_prods_merged.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max', 'sum']})
```

### - Results show that loyal customers spend the least on their orders on average than regular and new customers. New customers spend the most on average.

## 4.8.6 The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria: 
### a) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” 
### b) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create avg_price column using transform()
ords_prods_merged['avg_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Creating spending_flag column using loc()
ords_prods_merged.loc[ords_prods_merged['avg_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merged.loc[ords_prods_merged['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Removing row limits
pd.options.display.max_rows = None
```

```python
# Check output
ords_prods_merged[['user_id', 'avg_price', 'spending_flag']].head(100)
```

```python
# Check frequency of spending_flag
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

## 4.8.7 In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
### a) If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
### b) If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
### c) If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create med_days column using transform()
ords_prods_merged['med_days'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating frequency_flag column using loc()
ords_prods_merged.loc[ords_prods_merged['med_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merged.loc[(ords_prods_merged['med_days'] > 10) & (ords_prods_merged['med_days'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['med_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check output
ords_prods_merged[['user_id', 'med_days', 'frequency_flag']].head(60)
```

```python
# Check frequency
ords_prods_merged['frequency_flag'].value_counts(dropna = False)
```
